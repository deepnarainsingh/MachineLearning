Applying PCA On MNIST dataset of hand written digits
--

```python
#Importing different packages

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pylab as pl
%pylab inline
%matplotlib inline
pl.gray();
```

```python
#pl.xticks([]), pl.yticks([]) and  pl.axis('off')
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
digits.images.shape
```

```python
X=digits.data
y=digits.target
```

```python
print y.shape
print X.shape
```

```python
pl.gray();
pl.matshow(X) 
pl.show() 
```

```python
#Scaling the dataset
X_std = StandardScaler().fit_transform(X)
```

```python
#Doing PCA

pca = PCA(n_components=2)
X_New=pca.fit_transform(X_std)
```

```python
print X_New.shape
```

```python
plt.scatter(X_New[:, 0], X_New[:, 1], c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar();
```
